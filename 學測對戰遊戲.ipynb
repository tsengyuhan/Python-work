# 學測對戰遊戲

將學測的五個科目帶入史巴克拳的五種拳法，並將原本每一回合定勝負的規則改成扣血的方式，變成兩人對戰遊戲

## 以下為相關設定
1.史巴克拳：https://zh.wikipedia.org/wiki/石头、剪子、布、蜥蜴、史波克

2.學測五科：國文/英文/數學/自然/社會
<img style="float: right;"  src="史巴克拳.png"  width="40%">

3.雙方各有60的血量

4.雙方五種拳的攻擊力會依照所輸入的五科成績進行比例分配：滿級分攻擊力為20，越低分攻擊力等比降低(四捨五入)

5.由電腦擔任其中一方，資料來源為老師上課使用的假學測資料，從中隨機挑選一位作為對手

6.五科目之相互關係(左邊之科目可攻擊右邊)：

國文-->數學＆自然

英文-->國文＆數學

數學-->自然＆社會

社會-->英文＆國文

自然-->社會＆英文


## 遊戲規則
1.雙方輸入各自五科成績

2.遊戲開始，各自輸入所出的拳

3.進行攻擊，輸的一方被扣血

4.新的回合，各自輸入所出的拳，直到有一方血量為0


```python
%matplotlib inline 

import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
import matplotlib as mpl
from IPython.display import clear_output#從IPython.display 匯入一個特定的函式
import time #python標準的時間套件
mpl.rc('font', family='Noto Sans CJK TC')

```

```python
print('玩家進行設定：')
print('請輸入學測五科級分')

attack = ['國文', '英文', '數學' ,'社會' , '自然']
player_1 = np.empty(5)
for i in range(5):#玩家1設定戰力
    player_1[i]=input("請設定"+attack[i]+"級分:")
    while (player_1[i]>15) | (player_1[i]<0):
        player_1[i]=input("請重新設定"+attack[i]+"級分:")

clear_output(wait=True)
print('玩家設定完成，資料載入中請稍等.....')

player_2_df = pd.read_csv('https://bitly.com/gradescsv')#對手戰力設定：從表格中隨機選一人的學測成績
player_2_cho = player_2_df.iloc[random.randint(0,(len(player_2_df))-1)]
player_2 = [player_2_cho['國文'],player_2_cho['英文'],player_2_cho['數學'],player_2_cho['自然'],player_2_cho['社會']]

for x in range(5):#依照級分設定不同拳的攻擊力
    player_1[x] = round(20*(player_1[x]/15))
    player_2[x] = round(20*(player_2[x]/15))
        
print('玩家設定完成')
print('～！對戰開始！～')

hp_1=60
hp_2=60


while (hp_1>0) & (hp_2>0):#遊戲開始，直到一方血量為零結束
    print('玩家HP:'+str(hp_1))
    print('對手HP:'+str(hp_2))
    
    print('請出拳：請輸入科目')
    play_1_atc = input()
    if play_1_atc == '國文':
        play_1_atc = 0
    elif play_1_atc == '英文':
        play_1_atc = 1
    elif play_1_atc == '數學':
        play_1_atc = 2
    elif play_1_atc == '社會':
        play_1_atc = 3
    elif play_1_atc == '自然':
        play_1_atc = 4
        
   
    play_2_atc = random.randint(0,4)
    print('對手出：'+attack[play_2_atc])
    
    if(play_1_atc == 0): #出國文
        if(play_2_atc == 0):
            print('平手')
                  
        elif((play_2_atc == 1)|(play_2_atc == 3)):
            print('遭受攻擊！HP-' + str(player_2[play_2_atc]))
            hp_1 = hp_1 - player_2[play_2_atc]
                  
        elif((play_2_atc == 2)|(play_2_atc == 4)):
            print('發動攻擊！對方HP-' + str(player_1[play_1_atc]))
            hp_2 = hp_2 - player_1[play_1_atc]

    elif(play_1_atc == 1): #出英文
        if(play_2_atc == 1):
            print('平手')
                  
        elif((play_2_atc == 3)|(play_2_atc == 4)):
            print('遭受攻擊！HP-' + str(player_2[play_2_atc]))
            hp_1 = hp_1 - player_2[play_2_atc]
                  
        elif((play_2_atc == 0)|(play_2_atc == 2)):
            print('發動攻擊！對方HP-' + str(player_1[play_1_atc]))
            hp_2 = hp_2 - player_1[play_1_atc]
    
    elif(play_1_atc == 2): #出數學
        if(play_2_atc == 2):
            print('平手')
                  
        elif((play_2_atc == 0)|(play_2_atc == 1)):
            print('遭受攻擊！HP-' + str(player_2[play_2_atc]))
            hp_1 = hp_1 - player_2[play_2_atc]
                  
        elif((play_2_atc == 3)|(play_2_atc == 4)):
            print('發動攻擊！對方HP-' + str(player_1[play_1_atc]))
            hp_2 = hp_2 - player_1[play_1_atc]
            
    elif(play_1_atc == 3):#出社會
        if(play_2_atc == 3):
            print('平手')
                  
        elif((play_2_atc == 2)|(play_2_atc == 4)):
            print('遭受攻擊！HP-' + str(player_2[play_2_atc]))
            hp_1 = hp_1 - player_2[play_2_atc]
                  
        elif((play_2_atc == 0)|(play_2_atc == 1)):
            print('發動攻擊！對方HP-' + str(player_1[play_1_atc]))
            hp_2 = hp_2 - player_1[play_1_atc]
        
    elif(play_1_atc == 4):#出自然
        if(play_2_atc == 4):
            print('平手')
                  
        elif((play_2_atc == 0)|(play_2_atc == 2)):
            print('遭受攻擊！HP-' + str(player_2[play_2_atc]))
            hp_1 = hp_1 - player_2[play_2_atc]
                  
        elif((play_2_atc == 1)|(play_2_atc == 3)):
            print('發動攻擊！對方HP-' + str(player_1[play_1_atc]))
            hp_2 = hp_2 - player_1[play_1_atc]
            
    time.sleep(1.5)
    clear_output(wait=True)
    
if(hp_1<=0): #宣布遊戲結果，用小動畫呈現
    sp=0 #前面的空白數量
    emotion = '~GAME இдஇ OVER~'
    for i in range(30):
        k = (-1)**np.random.randint(1,3)#隨機左右移動，隨機大於等於1，小於3
        sp = sp+k
        if sp<0:#限制左右移動範圍
            sp=0
        elif sp>50:
            sp=50
        
        clear_output(wait=True)#清除每次顯示
        print(' '*sp + emotion)
        time.sleep(0.1)#每次間隔0.1 sec再繼續動作
else:
    sp=0 #前面的空白數量
    emotion = '~YOU ヽ(●´∀`●)ﾉ WIN~'
    for i in range(30):
        k = (-1)**np.random.randint(1,3)#隨機左右移動，隨機大於等於1，小於3
        sp = sp+k
        if sp<0:#限制左右移動範圍
            sp=0
        elif sp>50:
            sp=50
        clear_output(wait=True)#清除每次顯示
        print(' '*sp + emotion)
        time.sleep(0.1)#每次間隔0.1 sec再繼續動作
   
```
