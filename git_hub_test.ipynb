```python
df = pd.read_csv('http://stats.moe.gov.tw/files/detail/107/107_student.csv')
```

```python
df.head()#標題列＆第一列有問題
```

```python
df.loc[1] #第一列
```

```python
colname = df.loc[1].values
```

```python
df = df[2:] #從第二列開始
```

```python
df.head()
```

```python
df.columns = colname
```

```python
df.head()
```

```python
colname
```

```python
undergraduate = df[(df['日間∕進修別'] == 'D 日') & (df['等級別']== 'B 學士' ) ]
```

```python
undergraduate.head() #日間大學部
```

```python
#view：連回原來的資料集，改動就會改到原本的資料
#copy：複製一份，改動不會改到原來的
```

```python
df2 = undergraduate.loc[:, '學校名稱':'女生計'].copy()#學校名稱到女生數量間的資料
```

```python
df2.head()
```

```python
import locale#根據編碼做不同處理

#atof 把數字變成浮點數
```

```python
locale.setlocale(locale.LC_NUMERIC, "")
```

```python
df2['男生計'] = df2['男生計'].apply(locale.atof)
```

```python
df2['女生計'] = df2['女生計'].apply(locale.atof)
```

```python
df2['女男比例'] = df2.女生計/df2.男生計 #在資料集裡增加一欄男女比
```

```python
df2.head()
```

```python
df3 = df2.sort_values(by = ['女男比例'], ascending=False) #選定欄位，高排到低
```

```python
df3.head()
```

```python
df3[:20].plot.bar(x='學校名稱', y=['女男比例'])#畫出前二十名女男比例的長條圖
```
