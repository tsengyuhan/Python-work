```python
df = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=6283&md5_url=f62549441aa94775237b612eb0ef5464')
df
```

```python
df = df.loc[:,['學年度','男']]
```

```python
colnames = [ '學年度','男15歲','男6歲','成長率%']
df_all = pd.DataFrame(columns = colnames)
```

```python
for i in range(96,107):
    df_all.loc[(i-95),'學年度'] = i
    
df_all
```

```python
list_max= [ ]
list_min= [ ]
for x in range (96,107):
    a = df[df['學年度'] == x]
    list_max.append (max(a['男']))
    list_min.append (min(a['男']))
```

```python
list_max
```

```python
list_min
```

```python
df_all['男15歲']=list_max
df_all['男6歲']=list_min
```

```python
df_all
```

```python
df_all['成長率%'] = df_all['男15歲']/df_all['男6歲']*100
```

```python
df_all
```

```python
df_all.plot(x='學年度', y='成長率%')
```
